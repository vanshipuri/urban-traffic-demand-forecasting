```python
import pandas as pd
import numpy as np
import lightgbm as lgb
import pygeohash as pgh
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.preprocessing import OrdinalEncoder
import warnings

warnings.filterwarnings('ignore')

class TimestampExtractor(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None): return self
    def transform(self, X):
        X = X.copy()
        if "timestamp" in X.columns:
            parts = X["timestamp"].astype(str).str.split(":", expand=True)
            X["hour"] = parts[0].astype(int)
            X["minute"] = parts[1].astype(int)
            X["time_slot"] = X["hour"] * 4 + X["minute"] // 15
            X.drop(columns=["timestamp"], inplace=True)
        return X

class DayBinarizer(BaseEstimator, TransformerMixin):
    def __init__(self, column="day"): self.column = column
    def fit(self, X, y=None):
        self.min_val_ = X[self.column].min()
        return self
    def transform(self, X):
        X = X.copy()
        if self.column in X.columns:
            X[self.column] = (X[self.column] - self.min_val_).astype(int)
        return X

class CyclicalEncoder(BaseEstimator, TransformerMixin):
    def __init__(self, columns_periods=None, drop_original=True):
        self.columns_periods = columns_periods or {"hour": 24, "minute": 60}
        self.drop_original = drop_original
    def fit(self, X, y=None): return self
    def transform(self, X):
        X = X.copy()
        for col, period in self.columns_periods.items():
            if col in X.columns:
                X[f"{col}_sin"] = np.sin(2 * np.pi * X[col] / period)
                X[f"{col}_cos"] = np.cos(2 * np.pi * X[col] / period)
        if self.drop_original:
            cols_to_drop = [c for c in self.columns_periods if c in X.columns]
            X.drop(columns=cols_to_drop, inplace=True)
        return X

class BinaryEncoder(BaseEstimator, TransformerMixin):
    def __init__(self, mapping=None):
        self.mapping = mapping or {
            "LargeVehicles": {"Not Allowed": 0, "Allowed": 1},
            "Landmarks": {"No": 0, "Yes": 1},
        }
    def fit(self, X, y=None): return self
    def transform(self, X):
        X = X.copy()
        for col, m in self.mapping.items():
            if col in X.columns:
                X[col] = X[col].map(m).fillna(0)
        return X

class ExtraTreesMICEImputer(BaseEstimator, TransformerMixin):
    def __init__(self, random_state=42):
        self.random_state = random_state
        self.imputer = IterativeImputer(
            estimator=ExtraTreesRegressor(n_estimators=30, max_depth=8, random_state=self.random_state, n_jobs=-1),
            max_iter=5,
            random_state=self.random_state
        )
        self.oe = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=np.nan)
        self.cat_cols = ['RoadType', 'Weather']

    def fit(self, X, y=None):
        X_fit = X.copy()
        for col in self.cat_cols:
            if col in X_fit.columns and X_fit[col].dtype == 'object':
                X_fit[col] = X_fit[col].astype(str)
        X_fit[self.cat_cols] = self.oe.fit_transform(X_fit[self.cat_cols])

        self.num_cols = X_fit.select_dtypes(include=np.number).columns.tolist()
        self.imputer.fit(X_fit[self.num_cols])
        return self

    def transform(self, X):
        X_tf = X.copy()
        for col in self.cat_cols:
            if col in X_tf.columns and X_tf[col].dtype == 'object':
                X_tf[col] = X_tf[col].astype(str)
        X_tf[self.cat_cols] = self.oe.transform(X_tf[self.cat_cols])

        imputed_array = self.imputer.transform(X_tf[self.num_cols])
        X_tf[self.num_cols] = imputed_array

        X_tf[self.cat_cols] = np.round(X_tf[self.cat_cols])
        X_tf[self.cat_cols] = self.oe.inverse_transform(X_tf[self.cat_cols])
        return X_tf

class CategoricalCaster(BaseEstimator, TransformerMixin):
    # Geohash added back to native categories to utilize LightGBM Fisher Partitioning
    def __init__(self, columns=None):
        self.columns = columns or ['geohash', 'RoadType', 'Weather']
    def fit(self, X, y=None): return self
    def transform(self, X):
        X = X.copy()
        for col in self.columns:
            if col in X.columns:
                X[col] = X[col].astype('category')
        return X

class ColumnDropper(BaseEstimator, TransformerMixin):
    def __init__(self, columns=None): self.columns = columns or ["Index"]
    def fit(self, X, y=None): return self
    def transform(self, X):
        X = X.copy()
        cols_to_drop = [c for c in self.columns if c in X.columns]
        return X.drop(columns=cols_to_drop)

class TrafficPreprocessor:
    def __init__(self):
        self._build_steps()

    def _build_steps(self):
        self.steps = [
            ("drop_index", ColumnDropper(columns=["Index"])),
            ("timestamp", TimestampExtractor()),
            ("day_bin", DayBinarizer(column="day")),
            ("mice_impute", ExtraTreesMICEImputer(random_state=42)),
            ("cyclical", CyclicalEncoder(columns_periods={"hour": 24, "minute": 60}, drop_original=False)),
            ("binary", BinaryEncoder()),
            ("cat_caster", CategoricalCaster(columns=["geohash", "RoadType", "Weather"])),
        ]

    def fit_transform(self, X, y=None):
        Xt = X.copy()
        for name, step in self.steps:
            if hasattr(step, "fit_transform"):
                Xt = step.fit_transform(Xt, y)
            else:
                step.fit(Xt, y)
                Xt = step.transform(Xt)
        return Xt

    def transform(self, X):
        Xt = X.copy()
        for _, step in self.steps: Xt = step.transform(Xt)
        return Xt

# --- DATA LOADING & FEATURE ENGINEERING ---
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')

print("Engineering Lag and Spatial Features...")

def parse_time_robustly(df):
    parsed_dt = pd.to_datetime(df['timestamp'], format='%H:%M', errors='coerce')
    parsed_mins_str = parsed_dt.dt.hour * 60 + parsed_dt.dt.minute
    parsed_mins_num = pd.to_numeric(df['timestamp'], errors='coerce')
    return parsed_mins_str.fillna(parsed_mins_num)

train_df['total_minutes'] = parse_time_robustly(train_df)
test_df['total_minutes'] = parse_time_robustly(test_df)

train_df['hour'] = train_df['total_minutes'] // 60
test_df['hour'] = test_df['total_minutes'] // 60
train_df['day_section'] = train_df['hour'] // 4
test_df['day_section'] = test_df['hour'] // 4

demand_lookup = train_df.set_index(['day', 'geohash', 'total_minutes'])['demand'].to_dict()
def get_lag_demand(row, is_test=False):
    target_day = 48 if is_test else row['day'] - 1
    return demand_lookup.get((target_day, row['geohash'], row['total_minutes']), np.nan)

train_df['lag_1_demand'] = train_df.apply(lambda x: get_lag_demand(x, False), axis=1)
test_df['lag_1_demand'] = test_df.apply(lambda x: get_lag_demand(x, True), axis=1)

median_demand = train_df['demand'].median()
train_df['lag_1_demand'] = train_df['lag_1_demand'].fillna(median_demand)
test_df['lag_1_demand'] = test_df['lag_1_demand'].fillna(median_demand)

train_df['latitude'] = train_df['geohash'].apply(lambda x: pgh.decode(x)[0] if pd.notnull(x) else np.nan)
train_df['longitude'] = train_df['geohash'].apply(lambda x: pgh.decode(x)[1] if pd.notnull(x) else np.nan)
test_df['latitude'] = test_df['geohash'].apply(lambda x: pgh.decode(x)[0] if pd.notnull(x) else np.nan)
test_df['longitude'] = test_df['geohash'].apply(lambda x: pgh.decode(x)[1] if pd.notnull(x) else np.nan)

train_df.drop(columns=['total_minutes'], inplace=True)
test_df.drop(columns=['total_minutes'], inplace=True)

loc_stats = train_df.groupby('geohash')['Temperature'].agg(['mean', 'std']).reset_index()
loc_stats.rename(columns={'mean': 'temp_loc_mean', 'std': 'temp_loc_std'}, inplace=True)
time_stats = train_df.groupby('hour')['demand'].agg(['mean']).reset_index()
time_stats.rename(columns={'mean': 'demand_hour_mean'}, inplace=True)
section_stats = train_df.groupby('day_section')['demand'].agg(['mean']).reset_index()
section_stats.rename(columns={'mean': 'demand_section_mean'}, inplace=True)

for df in [train_df, test_df]:
    df = df.merge(loc_stats, on='geohash', how='left')
    df = df.merge(time_stats, on='hour', how='left')
    df = df.merge(section_stats, on='day_section', how='left')

# --- MODEL TRAINING ---
target_col = 'demand'
features = [c for c in train_df.columns if c != target_col]
cat_cols = ['geohash', 'RoadType', 'Weather']

# RESTORED: The Optuna hyper-parameters for high capacity
best_lgb_params = {
    'objective': 'regression',
    'metric': 'rmse',
    'n_estimators': 1500,
    'learning_rate': 0.0439736,
    'max_depth': 14,
    'num_leaves': 113,
    'feature_fraction': 0.6032,
    'subsample': 0.9525,
    'min_child_samples': 44,
    'reg_alpha': 0.0086,
    'reg_lambda': 0.0318,
    'random_state': 42,
    'verbosity': -1,
    'n_jobs': -1
}

kf = KFold(n_splits=5, shuffle=True, random_state=42)

oof_lag = np.zeros(len(train_df))
oof_global = np.zeros(len(train_df))
test_lag = np.zeros(len(test_df))
test_global = np.zeros(len(test_df))

print("\nStarting Unleashed Pipeline with MICE & Dual-Blend...")

for fold, (train_idx, val_idx) in enumerate(kf.split(train_df)):
    print(f"\n--- FOLD {fold+1} ---")

    X_tr_raw = train_df.iloc[train_idx][features]
    y_tr = train_df.iloc[train_idx][target_col]
    X_va_raw = train_df.iloc[val_idx][features]
    y_va = train_df.iloc[val_idx][target_col]

    preprocessor = TrafficPreprocessor()
    X_tr_proc_all = preprocessor.fit_transform(X_tr_raw, y_tr)
    X_va_proc_all = preprocessor.transform(X_va_raw)
    X_test_proc_all = preprocessor.transform(test_df[features])

    pipeline_features = list(X_tr_proc_all.columns)
    lag_features_fold = pipeline_features.copy()
    global_features_fold = [f for f in pipeline_features if f != 'lag_1_demand']
    global_cat_cols = [c for c in cat_cols if c in global_features_fold]

    print("  Training Lag Model...")
    model_lag = lgb.LGBMRegressor(**best_lgb_params)
    model_lag.fit(
        X_tr_proc_all[lag_features_fold], y_tr,
        eval_set=[(X_va_proc_all[lag_features_fold], y_va)],
        categorical_feature=cat_cols,
        callbacks=[lgb.early_stopping(150, verbose=False)]
    )
    oof_lag[val_idx] = model_lag.predict(X_va_proc_all[lag_features_fold])
    test_lag += model_lag.predict(X_test_proc_all[lag_features_fold]) / 5

    print("  Training Global Model...")
    model_glob = lgb.LGBMRegressor(**best_lgb_params)
    model_glob.fit(
        X_tr_proc_all[global_features_fold], y_tr,
        eval_set=[(X_va_proc_all[global_features_fold], y_va)],
        categorical_feature=global_cat_cols,
        callbacks=[lgb.early_stopping(150, verbose=False)]
    )
    oof_global[val_idx] = model_glob.predict(X_va_proc_all[global_features_fold])
    test_global += model_glob.predict(X_test_proc_all[global_features_fold]) / 5

final_oof = (0.80 * oof_lag) + (0.20 * oof_global)
final_test = (0.80 * test_lag) + (0.20 * test_global)

oof_r2 = r2_score(train_df[target_col], final_oof)
competition_score = max(0, 100 * oof_r2)

print("\n" + "="*45)
print(f"MICE Unleashed Blend R2:            {oof_r2:.5f}")
print(f"Final Estimated Competition Score:  {competition_score:.2f}")
print("="*45)

submission = pd.DataFrame({'Index': test_df['Index'], 'demand': final_test})
submission['demand'] = submission['demand'].clip(lower=0)
submission.to_csv('submission_unleashed_blend.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import lightgbm as lgb
import pygeohash as pgh
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.preprocessing import OrdinalEncoder
from sklearn.cluster import KMeans
import warnings

warnings.filterwarnings('ignore')

# =========================================================
# 1. OOP PREPROCESSING CLASSES
# =========================================================
class TimestampExtractor(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None): return self
    def transform(self, X):
        X = X.copy()
        if "timestamp" in X.columns:
            parts = X["timestamp"].astype(str).str.split(":", expand=True)
            X["hour"] = parts[0].astype(int)
            X["minute"] = parts[1].astype(int)
            X["time_slot"] = X["hour"] * 4 + X["minute"] // 15
            X.drop(columns=["timestamp"], inplace=True)
        return X

class DayBinarizer(BaseEstimator, TransformerMixin):
    def __init__(self, column="day"): self.column = column
    def fit(self, X, y=None):
        self.min_val_ = X[self.column].min()
        return self
    def transform(self, X):
        X = X.copy()
        if self.column in X.columns:
            X[self.column] = (X[self.column] - self.min_val_).astype(int)
        return X

class CyclicalEncoder(BaseEstimator, TransformerMixin):
    def __init__(self, columns_periods=None, drop_original=True):
        self.columns_periods = columns_periods or {"hour": 24, "minute": 60}
        self.drop_original = drop_original
    def fit(self, X, y=None): return self
    def transform(self, X):
        X = X.copy()
        for col, period in self.columns_periods.items():
            if col in X.columns:
                X[f"{col}_sin"] = np.sin(2 * np.pi * X[col] / period)
                X[f"{col}_cos"] = np.cos(2 * np.pi * X[col] / period)
        if self.drop_original:
            cols_to_drop = [c for c in self.columns_periods if c in X.columns]
            X.drop(columns=cols_to_drop, inplace=True)
        return X

class BinaryEncoder(BaseEstimator, TransformerMixin):
    def __init__(self, mapping=None):
        self.mapping = mapping or {
            "LargeVehicles": {"Not Allowed": 0, "Allowed": 1},
            "Landmarks": {"No": 0, "Yes": 1},
        }
    def fit(self, X, y=None): return self
    def transform(self, X):
        X = X.copy()
        for col, m in self.mapping.items():
            if col in X.columns:
                X[col] = X[col].map(m).fillna(0)
        return X

class ExtraTreesMICEImputer(BaseEstimator, TransformerMixin):
    def __init__(self, random_state=42):
        self.random_state = random_state
        self.imputer = IterativeImputer(
            estimator=ExtraTreesRegressor(n_estimators=30, max_depth=8, random_state=self.random_state, n_jobs=-1),
            max_iter=5,
            random_state=self.random_state
        )
        self.oe = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=np.nan)
        self.cat_cols = ['RoadType', 'Weather']

    def fit(self, X, y=None):
        X_fit = X.copy()
        for col in self.cat_cols:
            if col in X_fit.columns and X_fit[col].dtype == 'object':
                X_fit[col] = X_fit[col].astype(str)
        X_fit[self.cat_cols] = self.oe.fit_transform(X_fit[self.cat_cols])

        self.num_cols = X_fit.select_dtypes(include=np.number).columns.tolist()
        self.imputer.fit(X_fit[self.num_cols])
        return self

    def transform(self, X):
        X_tf = X.copy()
        for col in self.cat_cols:
            if col in X_tf.columns and X_tf[col].dtype == 'object':
                X_tf[col] = X_tf[col].astype(str)
        X_tf[self.cat_cols] = self.oe.transform(X_tf[self.cat_cols])

        imputed_array = self.imputer.transform(X_tf[self.num_cols])
        X_tf[self.num_cols] = imputed_array

        X_tf[self.cat_cols] = np.round(X_tf[self.cat_cols])
        X_tf[self.cat_cols] = self.oe.inverse_transform(X_tf[self.cat_cols])
        return X_tf

class OneHotEncoderTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, columns=None):
        self.columns = columns or ["Weather", "RoadType"]
    def fit(self, X, y=None):
        self.dummies_columns_ = {}
        for col in self.columns:
            if col in X.columns:
                dummies = pd.get_dummies(X[col], prefix=col, dtype=int)
                self.dummies_columns_[col] = dummies.columns.tolist()
        return self
    def transform(self, X):
        X = X.copy()
        for col, dummy_cols in self.dummies_columns_.items():
            if col in X.columns:
                dummies = pd.get_dummies(X[col], prefix=col, dtype=int)
                for dc in dummy_cols:
                    if dc not in dummies.columns:
                        dummies[dc] = 0
                dummies = dummies[dummy_cols]
                X = pd.concat([X.drop(columns=[col]), dummies], axis=1)
        return X

class CategoricalCaster(BaseEstimator, TransformerMixin):
    def __init__(self, columns=None):
        self.columns = columns or ['geohash', 'spatial_cluster']
    def fit(self, X, y=None): return self
    def transform(self, X):
        X = X.copy()
        for col in self.columns:
            if col in X.columns:
                X[col] = X[col].astype('category')
        return X

class ColumnDropper(BaseEstimator, TransformerMixin):
    def __init__(self, columns=None): self.columns = columns or ["Index"]
    def fit(self, X, y=None): return self
    def transform(self, X):
        X = X.copy()
        cols_to_drop = [c for c in self.columns if c in X.columns]
        return X.drop(columns=cols_to_drop)

class TrafficPreprocessor:
    def __init__(self):
        self._build_steps()
    def _build_steps(self):
        self.steps = [
            ("drop_index", ColumnDropper(columns=["Index"])),
            ("timestamp", TimestampExtractor()),
            ("day_bin", DayBinarizer(column="day")),
            ("mice_impute", ExtraTreesMICEImputer(random_state=42)),
            ("cyclical", CyclicalEncoder(columns_periods={"hour": 24, "minute": 60}, drop_original=False)),
            ("binary", BinaryEncoder()),
            ("onehot", OneHotEncoderTransformer(columns=["RoadType", "Weather"])),
            ("cat_caster", CategoricalCaster(columns=["geohash", "spatial_cluster"])),
        ]
    def fit_transform(self, X, y=None):
        Xt = X.copy()
        for name, step in self.steps:
            if hasattr(step, "fit_transform"):
                Xt = step.fit_transform(Xt, y)
            else:
                step.fit(Xt, y)
                Xt = step.transform(Xt)
        return Xt
    def transform(self, X):
        Xt = X.copy()
        for _, step in self.steps: Xt = step.transform(Xt)
        return Xt

# =========================================================
# 2. FEATURE ENGINEERING (Lag + Spatial K-Means)
# =========================================================
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')

print("Engineering Core Features & Spatial Clusters...")

def parse_time_robustly(df):
    parsed_dt = pd.to_datetime(df['timestamp'], format='%H:%M', errors='coerce')
    parsed_mins_str = parsed_dt.dt.hour * 60 + parsed_dt.dt.minute
    parsed_mins_num = pd.to_numeric(df['timestamp'], errors='coerce')
    return parsed_mins_str.fillna(parsed_mins_num)

train_df['total_minutes'] = parse_time_robustly(train_df)
test_df['total_minutes'] = parse_time_robustly(test_df)

train_df['hour'] = train_df['total_minutes'] // 60
test_df['hour'] = test_df['total_minutes'] // 60
train_df['day_section'] = train_df['hour'] // 4
test_df['day_section'] = test_df['hour'] // 4

demand_lookup = train_df.set_index(['day', 'geohash', 'total_minutes'])['demand'].to_dict()
def get_lag_demand(row, is_test=False):
    target_day = 48 if is_test else row['day'] - 1
    return demand_lookup.get((target_day, row['geohash'], row['total_minutes']), np.nan)

train_df['lag_1_demand'] = train_df.apply(lambda x: get_lag_demand(x, False), axis=1)
test_df['lag_1_demand'] = test_df.apply(lambda x: get_lag_demand(x, True), axis=1)

median_demand = train_df['demand'].median()
train_df['lag_1_demand'] = train_df['lag_1_demand'].fillna(median_demand)
test_df['lag_1_demand'] = test_df['lag_1_demand'].fillna(median_demand)

train_df['latitude'] = train_df['geohash'].apply(lambda x: pgh.decode(x)[0] if pd.notnull(x) else np.nan)
train_df['longitude'] = train_df['geohash'].apply(lambda x: pgh.decode(x)[1] if pd.notnull(x) else np.nan)
test_df['latitude'] = test_df['geohash'].apply(lambda x: pgh.decode(x)[0] if pd.notnull(x) else np.nan)
test_df['longitude'] = test_df['geohash'].apply(lambda x: pgh.decode(x)[1] if pd.notnull(x) else np.nan)

kmeans = KMeans(n_clusters=40, random_state=42)
train_coords = train_df[['latitude', 'longitude']].fillna(train_df[['latitude', 'longitude']].mean())
test_coords = test_df[['latitude', 'longitude']].fillna(test_df[['latitude', 'longitude']].mean())

train_df['spatial_cluster'] = kmeans.fit_predict(train_coords)
test_df['spatial_cluster'] = kmeans.predict(test_coords)

train_df.drop(columns=['total_minutes'], inplace=True)
test_df.drop(columns=['total_minutes'], inplace=True)

loc_stats = train_df.groupby('geohash')['Temperature'].agg(['mean', 'std']).reset_index()
loc_stats.rename(columns={'mean': 'temp_loc_mean', 'std': 'temp_loc_std'}, inplace=True)
time_stats = train_df.groupby('hour')['demand'].agg(['mean']).reset_index()
time_stats.rename(columns={'mean': 'demand_hour_mean'}, inplace=True)
section_stats = train_df.groupby('day_section')['demand'].agg(['mean']).reset_index()
section_stats.rename(columns={'mean': 'demand_section_mean'}, inplace=True)
cluster_stats = train_df.groupby('spatial_cluster')['demand'].agg(['mean']).reset_index()
cluster_stats.rename(columns={'mean': 'demand_cluster_mean'}, inplace=True)

for df in [train_df, test_df]:
    df = df.merge(loc_stats, on='geohash', how='left')
    df = df.merge(time_stats, on='hour', how='left')
    df = df.merge(section_stats, on='day_section', how='left')
    df = df.merge(cluster_stats, on='spatial_cluster', how='left')

# =========================================================
# 3. PHASE 1: BASE TRAINING (5-Fold CV)
# =========================================================
target_col = 'demand'
features = [c for c in train_df.columns if c != target_col]
cat_cols = ['geohash', 'spatial_cluster']

# UPDATED: Lower learning rate and higher n_estimators ceiling
best_lgb_params = {
    'objective': 'regression',
    'metric': 'rmse',
    'n_estimators': 4000,          # Increased to compensate for lower LR
    'learning_rate': 0.015,        # Dropped to 0.015 as requested
    'max_depth': 14,
    'num_leaves': 113,
    'feature_fraction': 0.6032,
    'subsample': 0.9525,
    'min_child_samples': 44,
    'reg_alpha': 0.0086,
    'reg_lambda': 0.0318,
    'random_state': 42,
    'verbosity': -1,
    'n_jobs': -1
}

kf = KFold(n_splits=5, shuffle=True, random_state=42)

oof_lag = np.zeros(len(train_df))
oof_global = np.zeros(len(train_df))
base_test_lag = np.zeros(len(test_df))
base_test_global = np.zeros(len(test_df))

print("\n[PHASE 1] Training Base Pipeline to Extract Pseudo-Labels...")

for fold, (train_idx, val_idx) in enumerate(kf.split(train_df)):
    print(f"  -> Fold {fold+1} / 5")

    X_tr_raw = train_df.iloc[train_idx][features]
    y_tr = train_df.iloc[train_idx][target_col]
    X_va_raw = train_df.iloc[val_idx][features]
    y_va = train_df.iloc[val_idx][target_col]

    preprocessor = TrafficPreprocessor()
    X_tr_proc_all = preprocessor.fit_transform(X_tr_raw, y_tr)
    X_va_proc_all = preprocessor.transform(X_va_raw)
    X_test_proc_all = preprocessor.transform(test_df[features])

    pipeline_features = list(X_tr_proc_all.columns)
    lag_features_fold = pipeline_features.copy()
    global_features_fold = [f for f in pipeline_features if f != 'lag_1_demand']
    global_cat_cols = [c for c in cat_cols if c in global_features_fold]

    model_lag = lgb.LGBMRegressor(**best_lgb_params)
    model_lag.fit(X_tr_proc_all[lag_features_fold], y_tr, eval_set=[(X_va_proc_all[lag_features_fold], y_va)], categorical_feature=cat_cols, callbacks=[lgb.early_stopping(150, verbose=False)])
    oof_lag[val_idx] = model_lag.predict(X_va_proc_all[lag_features_fold])
    base_test_lag += model_lag.predict(X_test_proc_all[lag_features_fold]) / 5

    model_glob = lgb.LGBMRegressor(**best_lgb_params)
    model_glob.fit(X_tr_proc_all[global_features_fold], y_tr, eval_set=[(X_va_proc_all[global_features_fold], y_va)], categorical_feature=global_cat_cols, callbacks=[lgb.early_stopping(150, verbose=False)])
    oof_global[val_idx] = model_glob.predict(X_va_proc_all[global_features_fold])
    base_test_global += model_glob.predict(X_test_proc_all[global_features_fold]) / 5

base_final_oof = (0.80 * oof_lag) + (0.20 * oof_global)
base_final_test = (0.80 * base_test_lag) + (0.20 * base_test_global)

oof_r2 = r2_score(train_df[target_col], base_final_oof)
print(f"  Base OOF R2: {oof_r2:.5f}")

# =========================================================
# 4. PHASE 2: PSEUDO-LABEL RETRAINING
# =========================================================
print("\n[PHASE 2] Generating Pseudo-Labels & Rebuilding Full Model...")

pseudo_test_df = test_df.copy()
pseudo_test_df[target_col] = base_final_test

combined_train_df = pd.concat([train_df, pseudo_test_df], ignore_index=True)

final_preprocessor = TrafficPreprocessor()
X_combined_proc = final_preprocessor.fit_transform(combined_train_df[features], combined_train_df[target_col])
X_final_test_proc = final_preprocessor.transform(test_df[features])

final_pipeline_features = list(X_combined_proc.columns)
final_lag_features = final_pipeline_features.copy()
final_global_features = [f for f in final_pipeline_features if f != 'lag_1_demand']
final_global_cat_cols = [c for c in cat_cols if c in final_global_features]

print("  Training Pseudo-Lag Model on 100% Data...")
final_model_lag = lgb.LGBMRegressor(**best_lgb_params)
final_model_lag.fit(X_combined_proc[final_lag_features], combined_train_df[target_col], categorical_feature=cat_cols)
pseudo_test_lag_preds = final_model_lag.predict(X_final_test_proc[final_lag_features])

print("  Training Pseudo-Global Model on 100% Data...")
final_model_glob = lgb.LGBMRegressor(**best_lgb_params)
final_model_glob.fit(X_combined_proc[final_global_features], combined_train_df[target_col], categorical_feature=final_global_cat_cols)
pseudo_test_glob_preds = final_model_glob.predict(X_final_test_proc[final_global_features])

ultimate_test_preds = (0.80 * pseudo_test_lag_preds) + (0.20 * pseudo_test_glob_preds)

print("\n" + "="*45)
print(f"Endgame Pipeline Completed Successfully.")
print(f"Base R2 Reference: {oof_r2:.5f}")
print("="*45)

submission = pd.DataFrame({'Index': test_df['Index'], 'demand': ultimate_test_preds})
submission['demand'] = submission['demand'].clip(lower=0)
submission.to_csv('submission_pseudo_ohe_blend_tuned.csv', index=False)
print("Saved Endgame Pseudo-Labeled predictions to 'submission_pseudo_ohe_blend_tuned.csv'")
```
